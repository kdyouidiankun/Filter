```python
### import two pkl files
import pickle
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
import matplotlib.ticker as ticker

### get numpy 
def get_numpy(file):
    with open(file, 'rb') as f:
        data = pickle.load(f)
    return np.array(data)

KFWs = get_numpy('NoMemKFWs.pkl')
LMSWs = get_numpy('NoMemLMSWs.pkl')

m, n = KFWs.shape[1], KFWs.shape[2]
```

```python
# Flatten the m x n dimensions into a single dimension
KFWs_flattened = KFWs.reshape(len(KFWs), m * n)
LMSWs_flattened = LMSWs.reshape(len(LMSWs), m * n)

# Plot a line for each element in the flattened arrays
plt.figure(figsize=(12, 8))

# Plot KFWs
for i in range(m * n):
    plt.plot(KFWs_flattened[:, i], label=f'KFWs[:, {i // n}, {i % n}]', linestyle='-', alpha=0.7)

# Plot LMSWs
for i in range(m * n):
    plt.plot(LMSWs_flattened[:, i], label=f'LMSWs[:, {i // n}, {i % n}]', linestyle='--', alpha=0.7)

plt.xlabel('Time Index')
plt.ylabel('Value')
plt.title('Line Plot for Each Element in KFWs and LMSWs')
plt.legend()
plt.show()
```
