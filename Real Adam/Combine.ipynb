```python
### read pngs (usually 4) in the same directory and combine them into 1 png
```

```python
def read_pngs():
    import os
    import cv2
    import numpy as np

    files = os.listdir()
    files = [f for f in files if f.endswith('.png')]
    files.sort()
    ### delete combined png
    if 'combined.png' in files:
        files.remove('combined.png')
        
    print(files)

    imgs = [cv2.imread(f, cv2.IMREAD_UNCHANGED) for f in files]
    imgs = [cv2.cvtColor(img, cv2.COLOR_BGR2RGB) for img in imgs]

    return imgs
```

```python
### combine images
imgs = read_pngs()
import numpy as np
import matplotlib.pyplot as plt

img1= np.concatenate(imgs[:2], axis=1)
img2= np.concatenate(imgs[2:], axis=1)
img = np.concatenate([img1, img2], axis=0)

plt.imshow(img)
plt.title('KF vs Adam, Conv, with Noise')
plt.savefig('combined.png', dpi=300)
```
