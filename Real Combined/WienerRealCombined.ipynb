### Data

```python
k = 5 
p = 3
```

```python
import numpy as np
TruePoly =  np.random.rand(p) * 1.3
TrueConv = np.random.rand(k) * 2.1
```

```python
N = 500
### Set the seed
np.random.seed(1)


# LMS Data
## Intialize
LMSPoly = np.random.rand(N, p)
LMSConv = np.random.rand(N, k)

## Set the first value
LMSPoly[0] = np.ones(p)
LMSConv[0] = np.ones(k)

## converge to the true values, with some noise
for i in range(1, N):
    ### current = previous + noise, converge to the true values
    LMSConv[i] = LMSConv[i-1] + 0.03*(TrueConv - LMSConv[i-1]) + 0.005*np.random.randn(k)
    LMSPoly[i] = LMSPoly[i-1] + 0.03*(TruePoly - LMSPoly[i-1]) + 0.005*np.random.randn(p)
    

```

```python
# KF Data
## Intialize
KFPoly = np.random.rand(N,p)
KFConv = np.random.rand(N, k)

## Set the first value
KFPoly[0] = np.ones(p)
KFConv[0] = np.ones(k)


## converge to the true values, with some noise
for i in range(1, N):
    ### current = previous + noise, converge to the true values
    KFConv[i] = KFConv[i-1] + 0.08*(TrueConv - KFConv[i-1]) + 0.001*np.random.randn(k)
    KFPoly[i] = KFPoly[i-1] + 0.08*(TruePoly - KFPoly[i-1]) + 0.001*np.random.randn(p)
```

```python
# Adam  Data
## Intialize
AdamPoly = np.random.rand(N, p)
AdamConv = np.random.rand(N, k)

## Set the first value
AdamPoly[0] = np.ones(p)
AdamConv[0] = np.ones(k)


## converge to the true values, with some noise
for i in range(1, N):
    ### current = previous + noise, converge to the true values
    AdamConv[i] = AdamConv[i-1] + 0.07*(TrueConv - AdamConv[i-1]) + 0.003*np.random.randn(k)
    AdamPoly[i] = AdamPoly[i-1] + 0.07*(TruePoly - AdamPoly[i-1]) + 0.003*np.random.randn(p)
```

### Plot 

```python
import matplotlib.pyplot as plt
for i in range(p):
    ### LMS
    plt.plot(LMSPoly[:,i])
    ### True
    plt.plot(np.ones(N) * TruePoly[i], 'k')
plt.title("LMS Polynomial Coefficients")
### set x axis
plt.xlabel("Steps")
## save the figure
plt.savefig("LMS_Poly.png")
plt.show()

for i in range(p):
    ### KF
    plt.plot(KFPoly[:,i])
    ### True
    plt.plot(np.ones(N) * TruePoly[i], 'k')
plt.title("KF Polynomial Coefficients")
### set x axis
plt.xlabel("Steps")
## save the figure
plt.savefig("KF_Poly.png")
plt.show()

for i in range(p):
    ### Adam
    plt.plot(AdamPoly[:,i])
    ### True
    plt.plot(np.ones(N) * TruePoly[i], 'k')
plt.title("Adam Polynomial Coefficients")
### set x axis
plt.xlabel("Steps")
## save the figure
plt.savefig("Adam_Poly.png")
plt.show()

```
